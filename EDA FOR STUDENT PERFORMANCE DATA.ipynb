### Handle Missing values

#2 options

- Remove the missing values
- Replace with specific values

```python
## drop missing values on studytimeweekly and absences

data.dropna(subset=["StudyTimeWeekly", "Absences"], inplace=True)
```

```python
data.isna().sum()
```

```python
parental_ed_mode = data["ParentalEducation"].mode()[0]
```

```python
parental_support_mode = data["ParentalSupport"].mode()[0]
```

```python
# replace missing values in ParentalEducation  and ParentalSupport  

data[["ParentalEducation", "ParentalSupport"]] = data[["ParentalEducation", "ParentalSupport"]].fillna({
    "ParentalEducation": parental_ed_mode,
    "ParentalSupport": parental_support_mode
})
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
gpa_avg = data["GPA"].median()
```

```python
gpa_avg
```

```python
data["GPA"] = data["GPA"].fillna(gpa_avg)
```

```python
# Step 1: Calculate Q1, Q3, and IQR
Q1 = data["Age"].quantile(0.25)
Q3 = data["Age"].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 3: Find outliers
outliers = data[(data["Age"] < lower_bound) | (data["Age"] > upper_bound)]

# Step 4: Count them
num_outliers = outliers.shape[0]

print(f"Number of outliers in 'Age': {num_outliers}")
```
