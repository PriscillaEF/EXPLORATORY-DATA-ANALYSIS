#### Smallest populations in 2022

```python
#top range columns based off 2022 population

df.sort_values(by="2022 Population").head()
```

```python
#largest to smallest by 2022 Population

df.sort_values(by="2022 Population", ascending =False).head(10)
```

China and India dominate

A small number of countries account for a large share of humanity

```python
#largest to smallest by World Population Percentage

df.sort_values(by="World Population Percentage", ascending =False).head(10)
```

#### Correlation Analysis

```python
#correlation btween numeric columns

df.corr(numeric_only=True)
```

Strong correlation between population across years

Density and area behave differently

Growth rate weakly correlated with total population

```python
#visualize it

sns.heatmap(df.corr(numeric_only=True), annot = True)

plt.rcParams['figure.figsize'] = (20,2)

plt.show
```

#### Grouping by continent to shift analysis to regional level

```python
#group columns by continent

df.groupby('Continent').mean(numeric_only=True).sort_values(by = "2022 Population", ascending=False)
```

Asia leads by a large margin

Africa shows strong population size and growth potential

```python
#Investigating Oceania

df[df['Continent'].str.contains('Oceania')]
```

Many small countries

Large landmass (Australia) but low population density

```python
df2 = df.groupby('Continent').mean(numeric_only=True).sort_values(by = "2022 Population", ascending=False)

df2
```

```python
#switch continent to column and column becomes index

df3=df2.transpose()
```
